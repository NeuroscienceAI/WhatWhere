```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

nmi_1oct = np.array([0.69, 0.65, 0.69, 0.68, 0.64, 0.67, 0.64, 0.63, 0.66, 0.58])
nmi_2oct = np.array([0.85, 0.82, 0.82, 0.81, 0.82, 0.83, 0.78, 0.83, 0.80, 0.60])
print(f"1 octave mean and sd= {np.mean(nmi_1oct):.2f} {np.std(nmi_1oct):.2f}")
print(f"2 octave mean and sd= {np.mean(nmi_2oct):.2f} {np.std(nmi_2oct):.2f}")
t_stat, p_val = ttest_rel(nmi_2oct, nmi_1oct)
print(f"t-statistic = {t_stat:.3f}")
print(f"p-value     = {p_val:.3e}")

arch_ids = np.arange(1, len(nmi_1oct) + 1)

fig, ax = plt.subplots(figsize=(4, 4))

ax.scatter(nmi_1oct, nmi_2oct)

for x, y, i in zip(nmi_1oct, nmi_2oct, arch_ids):
    ax.text(x, y, str(i), ha='center', va='bottom', fontsize=12)

min_val = min(nmi_1oct.min(), nmi_2oct.min())
max_val = max(nmi_1oct.max(), nmi_2oct.max())
ax.plot([min_val, max_val], [min_val, max_val], linestyle='--')

ax.set_xlabel("1 octave NMI")
ax.set_ylabel("2 octave NMI")
ax.set_title("NMI: 1 octave vs 2 octave")

padding = 0.02
ax.set_xlim(min_val - padding, max_val + padding)
ax.set_ylim(min_val - padding, max_val + padding)
ax.set_aspect('equal', adjustable='box')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('IHC3000_A1-10_1vs2oct_neighbor=100_NMI.pdf', dpi=300, bbox_inches='tight')
plt.show()

```

```python
r2_1oct = np.array([0.77, 0.70, 0.78, 0.73, 0.78, 0.77, 0.72, 0.73, 0.82, 0.76])
r2_2oct = np.array([0.74, 0.62, 0.76, 0.64, 0.68, 0.69, 0.62, 0.69, 0.71, 0.60])
print(f"1 octave mean and sd= {np.mean(r2_1oct):.2f} {np.std(r2_1oct):.2f}")
print(f"2 octave mean and sd= {np.mean(r2_2oct):.2f} {np.std(r2_2oct):.2f}")
t_stat, p_val = ttest_rel(r2_2oct, r2_1oct)
print(f"t-statistic = {t_stat:.3f}")
print(f"p-value     = {p_val:.3e}")

# architecture indices 110
arch_ids = np.arange(1, len(r2_1oct) + 1)

fig, ax = plt.subplots(figsize=(4, 4))

# scatter: 1 octave on X, 2 octave on Y
ax.scatter(r2_1oct, r2_2oct)

# label each dot with its architecture index
for x, y, i in zip(r2_1oct, r2_2oct, arch_ids):
    ax.text(x, y, str(i), ha='center', va='bottom', fontsize=12)

# y = x reference line
min_val = min(r2_1oct.min(), r2_2oct.min())
max_val = max(r2_1oct.max(), r2_2oct.max())
ax.plot([min_val, max_val], [min_val, max_val], linestyle='--')

ax.set_xlabel("1 octave $R^2$")
ax.set_ylabel("2 octave $R^2$")
ax.set_title("$R^2$: 1 octave vs 2 octave")

# make axes square and padded a bit
padding = 0.02
ax.set_xlim(min_val - padding, max_val + padding)
ax.set_ylim(min_val - padding, max_val + padding)
ax.set_aspect('equal', adjustable='box')

# remove top and right edges
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('IHC3000_A1-10_1vs2oct_neighbor=100_R2.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
