```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

nmi_800 = np.array([0.77, 0.75, 0.77, 0.78, 0.75, 0.78, 0.76, 0.73, 0.77, 0.76])
nmi_1400 = np.array([0.87, 0.83, 0.85, 0.87, 0.81, 0.85, 0.82, 0.82, 0.82, 0.81])
print(f"800Hz mean and sd= {np.mean(nmi_800):.2f} {np.std(nmi_800):.2f}")
print(f"1.4kHz mean and sd= {np.mean(nmi_1400):.2f} {np.std(nmi_1400):.2f}")
t_stat, p_val = ttest_rel(nmi_1400, nmi_800)
print(f"t-statistic = {t_stat:.3f}")
print(f"p-value     = {p_val:.3e}")

arch_ids = np.arange(1, len(nmi_800) + 1)

fig, ax = plt.subplots(figsize=(4, 4))

ax.scatter(nmi_800, nmi_1400)

for x, y, i in zip(nmi_800, nmi_1400, arch_ids):
    ax.text(x, y, str(i), ha='center', va='bottom', fontsize=12)

min_val = min(nmi_800.min(), nmi_1400.min())
max_val = max(nmi_800.max(), nmi_1400.max())
ax.plot([min_val, max_val], [min_val, max_val], linestyle='--')

ax.set_xlabel("800 Hz NMI")
ax.set_ylabel("1.4 kHz NMI")
ax.set_title("NMI: 800 Hz vs 1.4 kHz")

padding = 0.02
ax.set_xlim(min_val - padding, max_val + padding)
ax.set_ylim(min_val - padding, max_val + padding)
ax.set_aspect('equal', adjustable='box')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('IHC3000_A1-10_800vs1400Hz_neighbor=100_NMI.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
r2_800 = np.array([0.95, 0.99, 0.98, 0.94, 0.85, 0.90, 0.97, 0.86, 0.89, 0.91])
r2_1400 = np.array([0.74, 0.62, 0.76, 0.64, 0.68, 0.69, 0.62, 0.69, 0.71, 0.60])
print(f"800Hz mean and sd= {np.mean(r2_800):.2f} {np.std(r2_800):.2f}")
print(f"1.4kHz mean and sd= {np.mean(r2_1400):.2f} {np.std(r2_1400):.2f}")
t_stat, p_val = ttest_rel(r2_1400, r2_800)
print(f"t-statistic = {t_stat:.3f}")
print(f"p-value     = {p_val:.3e}")

# architecture indices 110
arch_ids = np.arange(1, len(r2_800) + 1)

fig, ax = plt.subplots(figsize=(4, 4))

# scatter: 1 octave on X, 2 octave on Y
ax.scatter(r2_800, r2_1400)

# label each dot with its architecture index
for x, y, i in zip(r2_800, r2_1400, arch_ids):
    ax.text(x, y, str(i), ha='center', va='bottom', fontsize=12)

# y = x reference line
min_val = min(r2_800.min(), r2_1400.min())
max_val = max(r2_800.max(), r2_1400.max())
ax.plot([min_val, max_val], [min_val, max_val], linestyle='--')

ax.set_xlabel("800 Hz $R^2$")
ax.set_ylabel("1.4 kHz $R^2$")
ax.set_title("$R^2$: 800 Hz vs 1.4 kHz")

# make axes square and padded a bit
padding = 0.02
ax.set_xlim(min_val - padding, max_val + padding)
ax.set_ylim(min_val - padding, max_val + padding)
ax.set_aspect('equal', adjustable='box')

# remove top and right edges
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('IHC3000_A1-10_800vs1400Hz_neighbor=100_R2.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
