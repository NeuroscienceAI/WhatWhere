```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

nmi_anechoic = np.array([1, 1, 1, 1, 0.99, 1, 0.99, 1, 1, 0.99])
nmi_reverb = np.array([1, 1, 1, 1, 0.99, 1, 0.99, 0.99, 1, 0.97])
print(f"anechoic mean and sd= {np.mean(nmi_anechoic):.2f} {np.std(nmi_anechoic):.2f}")
print(f"reverb mean and sd= {np.mean(nmi_reverb):.2f} {np.std(nmi_reverb):.2f}")
t_stat, p_val = ttest_rel(nmi_anechoic, nmi_reverb)
print(f"t-statistic = {t_stat:.3f}")
print(f"p-value     = {p_val:.3e}")

arch_ids = np.arange(1, len(nmi_reverb) + 1)

fig, ax = plt.subplots(figsize=(4, 4))

ax.scatter(nmi_anechoic, nmi_reverb)

for x, y, i in zip(nmi_anechoic, nmi_reverb, arch_ids):
    ax.text(x, y, str(i), ha='center', va='bottom', fontsize=12)

min_val = min(nmi_reverb.min(), nmi_anechoic.min())
max_val = max(nmi_reverb.max(), nmi_anechoic.max())
ax.plot([0.94, 1], [0.94, 1], linestyle='--')

ax.set_xlabel("anechoic NMI")
ax.set_ylabel("reverb NMI")
ax.set_title("NMI: anechoic vs reverb")

padding = 0.02
ax.set_xlim(0.95 - padding, 1 + padding)
ax.set_ylim(0.95 - padding, 1 + padding)
ax.set_aspect('equal', adjustable='box')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('IHC3000_A1-10_anechoic_vs_reverb_NMI_neighbor=20_NMI.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
r2_anechoic = np.array([0.68, 0.47, 0.76, 0.63, 0.49, 0.54, 0.75, 0.51, 0.44, 0.80])
r2_reverb = np.array([0.72, 0.52, 0.75, 0.48, 0.74, 0.63, 0.79, 0.70, 0.65, 0.86])
print(f"anechoic mean and sd= {np.mean(r2_anechoic):.2f} {np.std(r2_anechoic):.2f}")
print(f"reverb mean and sd= {np.mean(r2_reverb):.2f} {np.std(r2_reverb):.2f}")
t_stat, p_val = ttest_rel(r2_anechoic, r2_reverb)
print(f"t-statistic = {t_stat:.3f}")
print(f"p-value     = {p_val:.3e}")

# architecture indices 110
arch_ids = np.arange(1, len(r2_reverb) + 1)

fig, ax = plt.subplots(figsize=(4, 4))

# scatter: 1 octave on X, 2 octave on Y
ax.scatter(r2_anechoic, r2_reverb)

# label each dot with its architecture index
for x, y, i in zip(r2_anechoic, r2_reverb, arch_ids):
    ax.text(x, y, str(i), ha='center', va='bottom', fontsize=12)

# y = x reference line
min_val = min(r2_reverb.min(), r2_anechoic.min())
max_val = max(r2_reverb.max(), r2_anechoic.max())
ax.plot([min_val, max_val], [min_val, max_val], linestyle='--')

ax.set_xlabel("anechoic $R^2$")
ax.set_ylabel("reverb $R^2$")
ax.set_title("$R^2$: anechoic vs reverb")

# make axes square and padded a bit
padding = 0.02
ax.set_xlim(min_val - padding, max_val + padding)
ax.set_ylim(min_val - padding, max_val + padding)
ax.set_aspect('equal', adjustable='box')

# remove top and right edges
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('IHC3000_A1-10_anechoic_vs_reverb_neighbor=20_R2.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
